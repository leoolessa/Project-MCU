```python
sales_2009 = pd.read_csv('data/sales_2009.csv')
sales_2009
```

```python
def remove_columns(data):
    data.drop(['Units', 'Dollars', 'Issue'], axis=1, inplace=True)
    return data
```

```python
def add_date(data, date):
    data['Publish Date'] = date
    second_column = data.pop('Publish Date')
    data.insert(1, 'Publish Date', second_column)
    return date
```

```python
def remove_dolar(data):
    for c, r in data.iterrows():
        data.at[c, 'Price'] = r['Price'].replace('$', '')
    return data
```

```python
def change_column_type(data):
    for c, r in data.iterrows():
        try:
            data.at[c, 'Price'] = float(r['Price'])
        except ValueError:
            print(f"Could not convert value to float at index {c}: {r['Price']}")
    data['Price'] = data['Price'].astype(float)
    return data
```

```python
def change_comma_to_dot(data):
    for c, r in data.iterrows():
        data.at[c, 'Est. units'] = r['Est. units'].replace(',', '.')
    return data
```

```python
def change_column_type2(data):
    for c, r in data.iterrows():
        try:
            data.at[c, 'Est. units'] = float(r['Est. units'])
        except ValueError:
            print(f"Could not convert value to float at index {c}: {r['Est. units']}")
    data['Est. units'] = data['Est. units'].astype(float)
    return data
```

```python
def soma(data):
    for c, r in data.iterrows():
        data.at[c, 'Total'] = r['Price'] * r['Est. units']
    return data
```
