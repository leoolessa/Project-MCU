```python
sales_2009 = pd.read_csv('data/sales_2009.csv')
sales_2009
```

```python
def remove_dolar(data):
    for c, r in data.iterrows():
        data['Price'] = r['Price'].replace('$', '')
    return data
```

```python
def change_comma_to_dot(data):
    for c, r in data.iterrows():
        data['Est. units'] = r['Est. units'].replace(',', '.')
    return data
```

```python
change_comma_to_dot(sales_2009)
```

```python
def change_column_type(data):
    for c, r in data.iterrows():
        try:
            data.at[c, 'Est. units'] = float(r['Est. units'])
        except ValueError:
            print(f"Could not convert value to float at index {c}: {r['Est. units']}")
    data['Est. units'] = data['Est. units'].astype(float)
    return data
change_column_type(sales_2009)
```

```python
def soma(data):
    data['Total'] = data['Price'] * data['Est. units']
    return data
soma(sales_2009)
```
